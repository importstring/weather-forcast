```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# File path to the CSV file
filepath = '/Users/simon/Desktop/Areas/TKS/Focus 1 - RL/IRL-RL Application/Databases/Data/Toronto Weather Data/2024.csv'

# Read the CSV file
data = pd.read_csv(filepath)

# Select relevant numeric columns for correlation analysis
# (Adjust the column names as needed)
numeric_columns = [
    'Max Temp (C)', 'Min Temp (C)', 'Mean Temp (C)', 
    'Heat Deg Days (C)', 'Cool Deg Days (C)', 
    'Total Rain (mm)', 'Total Snow (cm)', 
    'Total Precip (mm)', 'Snow on Grnd (cm)', 
    'Spd of Max Gust (km/h)'
]

# Create a new DataFrame with only the numeric columns
numeric_data = data[numeric_columns]

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Create a heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)

# Add titles and labels
plt.title('Correlation Matrix of Climate Data', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
